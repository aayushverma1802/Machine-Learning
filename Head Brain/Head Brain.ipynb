```python
#Collecting the data
X=data["Head Size(cm^3)"].values
Y=data["Brain Weight(grams)"].values
```

```python
#Mead X and Y
mean_x=np.mean(X)
mean_y=np.mean(Y)

#Total number of values
n=len(X)

#Using the formula to calculate b1 and b2 
numer=0
denom=0
for i in range (n):
    numer +=(X[i] -mean_x)*(Y[i]-mean_y)
    denom +=(X[i]-mean_x)**2
b1=numer/denom
b0=mean_y-(b1*mean_x)
#Printing the Coefficient
print(b1,b0)
```
